```python

# calcular area y perimetro del circulo
# variables A,P,r
# formulas A = pi*r^2 y P = 2 * pi * r
# capturar el valor de r
import math
print("calcular area y perimetro del circulo")
r = float(input("ingrese el valor del radio: "))
A = math.pi* r**2
P = 2*math.pi*r
print("el area del circulo es: ",A)
print("el perimetro del circulo es: ",P)
print("valeria yaretzi NL = 32")
```
